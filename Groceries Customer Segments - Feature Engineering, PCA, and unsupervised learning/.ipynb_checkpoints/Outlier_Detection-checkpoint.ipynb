## Customer outlier detection 


In this code I want to get rid of the outliers in the grocery data set. I have written a separate file since I would like to discover different outlier detection algorithms. I am going to consider two methods I implement myself, and two who are already available in sci-kit:
    
*Self implemented*
1. [Tukey's method](https://sphweb.bumc.bu.edu/otlt/mph-modules/bs/bs704_summarizingdata/bs704_summarizingdata7.html) - also known as the IQR method. The Tukey’s method defines an outlier as those values of the data set that fall far from the central point, the median. One can think of plotting a boxplot and then find an outlier.
2. [Z-Scores](https://support.hach.com/ci/okcsFattach/get/1008007_4) Z-scores are a tool for determining outlying data based on data locations on graphs. Z-scores base this information on data distribution and using the standard deviation measurements of data to calculate outlier under the understanding that about 68% of measurements will be within one standard deviation of the mean and about 95% of measurements will be within two standard deviations of the mean.

*Sci-kit*
1. [Isolation Forest](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html) Isolation forest is an unsupervised learning algorithm for anomaly detection that works on the principle of isolating anomalies. The most common techniques employed for anomaly detection are based on the construction of a profile of what is “normal”: anomalies are reported as those instances in the dataset that do not conform to the normal profile. Isolation Forest uses a different approach: instead of trying to build a model of normal instances, it explicitly isolates anomalous points in the dataset. The main advantage of this approach is the possibility of exploiting sampling techniques to an extent that is not allowed to the profile-based methods, creating a very fast algorithm with a low memory demand.
2. [Elliptic Envelope](https://scikit-learn.org/stable/modules/generated/sklearn.covariance.EllipticEnvelope.html)The Elliptic Envelope method fits a multivariate gaussian distribution to the dataset. Use the contamination hyperparameter to specify the percentage of observations the algorithm will assign as outliers.

```python
spendings_boxcox=pd.read_pickle('DataFrames/spendings_boxcox')
spendings_log=pd.read_pickle('DataFrames/spendings_log')

Products=spendings_boxcox.columns
spendings_boxcox.head()
```

## Code verification and model selection
We have 6 independent features, I will approach the outlier detection problem six times as a one dimensional problem. After that we can drop the outliers. Let's first write the codes for 1 feature after which I loop over the others.

### Tukey's Method

Tukey's Method for outliers identification reads as follows: An outlier step is calculated as 1.5times the interquartile range (IQR). A data point with a feature that is beyond an outlier step outside of the IQR for that feature is considered an outlier.

```python
## First I calculate the first and third quartile
Q1 = np.percentile(spendings_boxcox['Milk'],25)
Q3 = np.percentile(spendings_boxcox['Milk'],75)

spendings_boxcox['Outlier_milk_IQR']=0 # For plotting

## We consider it an outlier if the step is 1.5 Q1-Q3:
change = (Q3-Q1) * 1.5
   
outlier=spendings_boxcox['Milk'][(spendings_boxcox['Milk'] <= Q1 - change) | (spendings_boxcox['Milk'] >= Q3 + change)]
print('Outliers at indices:', outlier.index.values)
spendings_boxcox['Outlier_milk_IQR'].iloc[outlier.index]=1
print('Verification using value counts: ')
spendings_boxcox['Outlier_milk_IQR'].value_counts()
```

```python
plt.figure()
sns.scatterplot(x=spendings_boxcox.index,y='Milk',data=spendings_boxcox,hue='Outlier_milk_IQR',style='Outlier_milk_IQR',c='green',palette='RdYlGn',alpha=1)
```

We see that the code selected two outliers. Let's see how the next does.

## Z-Values

This method defines outliers as data-points which are >3 standard deviations from the mean. We start by calculating the z-values defined as:
    $                z_i=\frac{x_i-\bar{x}}{\sigma_x}$

```python
spendings_boxcox['Outlier_milk_Z']=0 # For plotting

z_values=(spendings_boxcox['Milk']-spendings_boxcox['Milk'].mean())/spendings_boxcox['Milk'].std()
outliers=spendings_boxcox['Milk'][abs(z_values)>3]

print('Outliers at indices:', outlier.index.values)
spendings_boxcox['Outlier_milk_Z'].iloc[outlier.index]=1
print('Verification using value counts: ')
spendings_boxcox['Outlier_milk_Z'].value_counts()
```

```python
plt.figure()
sns.scatterplot(x=spendings_boxcox.index,y='Milk',data=spendings_boxcox,hue='Outlier_milk_Z',style='Outlier_milk_Z',c='green',palette='RdYlGn',alpha=1)
```

It can be seen that the same point are selected by this algoritm so that verifies both codes.

## Isolation Forest
This algo is already implemented in sci-kit and used from the library. This algoritm requires a threshcode, i.e. the algorithm considers an x percentage of the points as outliers. Based on the previous I decided to use a ratio of 1\%.

```python
from sklearn.ensemble import IsolationForest

spendings_boxcox['Outlier_milk_IsoFor']=1

IsoFor=IsolationForest(max_samples=440, random_state=101,contamination=.01)
IsoFor.fit(spendings_boxcox['Milk'].values.reshape(-1,1))
IsoFor_outlier = IsoFor.predict(spendings_boxcox['Milk'].values.reshape(-1,1))
spendings_boxcox['Outlier_milk_IsoFor']=spendings_boxcox['Outlier_milk_IsoFor']*IsoFor_outlier<1
plt.figure()
sns.scatterplot(x=spendings_boxcox.index,y='Milk',data=spendings_boxcox,hue='Outlier_milk_IsoFor',style='Outlier_milk_IsoFor',c='green',palette='RdYlGn',alpha=1)
```

We see that some more points are considerd to be outliers compared to the earlier algos. This is because I set it to be 1\%. The selected point match the selected points of IQR and z-score though.

## Full outlier detection
Now it's time to consider each of the six features independently and then remove the outliers from the original dataframe.

### IQR Full

```python
outliers_boxcox=[]
outliers_log=[]

for Product in Products:
    Q1_boxcox = np.percentile(spendings_boxcox[Product],25)
    Q3_boxcox = np.percentile(spendings_boxcox[Product],75)

    Q1_log = np.percentile(spendings_log[Product],25)
    Q3_log = np.percentile(spendings_log[Product],75)
    ## We consider it an outlier if the step is 1.5 Q1-Q3:
    change_boxcox = (Q3_boxcox-Q1_boxcox) * 1.5
    change_log = (Q3_log-Q1_log) * 1.5

    outlier_boxcox=spendings_boxcox[Product][(spendings_boxcox[Product] <= Q1_boxcox - change_boxcox) | (spendings_boxcox[Product] >= Q3_boxcox + change_boxcox)]
    outlier_log=spendings_log[Product][(spendings_log[Product] <= Q1_log - change_log) | (spendings_log[Product] >= Q3_log + change_log)]

    for i in range(0,len(outlier_boxcox)): ## Append the selected outliers for each feature to athe total list
        outliers_boxcox.append(outlier_boxcox.index[i])
    for i in range(0,len(outlier_log)): ## Append the selected outliers for each feature to athe total list
        outliers_log.append(outlier_log.index[i])
        
outliers_boxcox=np.array(outliers_boxcox)
outliers_boxcox=np.unique(outliers_boxcox) # select all uniques
print('Outlier indices boxcox: ',outliers_boxcox)
print('# of outliers boxcox: ',len(outliers_boxcox))

outliers_log=np.array(outliers_log)
outliers_log=np.unique(outliers_log) # select all uniques
print('Outlier indices log: ',outliers_log)
print('# of outliers log: ',len(outliers_log))
```

```python
spendings_log_IQR=pd.read_pickle('DataFrames/spendings_log')
spendings_log_IQR.drop(outliers_log, inplace=True)
print('So we should drop 42 of the 440 this if verified by the length of the remaining: ', len(spendings_log_IQR))
spendings_log_IQR.to_pickle('DataFrames/spendings_log_IQR')
```

### Isolation forest full

```python
outliers_boxcox=[]
outliers_log=[]

IsoFor_boxcox=IsolationForest(max_samples=440, random_state=101,contamination=.01)
IsoFor_log=IsolationForest(max_samples=440, random_state=101,contamination=.01)

for Product in Products:
    IsoFor_boxcox.fit(spendings_boxcox[Product].values.reshape(-1,1))
    IsoFor_outlier_boxcox = IsoFor_boxcox.predict(spendings_boxcox[Product].values.reshape(-1,1))
    outlier_boxcox=np.where(IsoFor_outlier_boxcox<0)
    
    IsoFor_log.fit(spendings_log[Product].values.reshape(-1,1))
    IsoFor_outlier_log= IsoFor_log.predict(spendings_log[Product].values.reshape(-1,1))
    outlier_log=np.where(IsoFor_outlier_log<0)

    for i in range(0,len(outlier_boxcox[0])):
        outliers_boxcox.append(outlier_boxcox[0][i])
        
    for i in range(0,len(outlier_log[0])):
        outliers_log.append(outlier_log[0][i])
        
outliers_boxcox=np.array(outliers_boxcox)
outliers_boxcox=np.unique(outliers_boxcox)
print('Outlier indices boxcox: ',outliers_boxcox)
print('# of outliers boxcox: ',len(outliers_boxcox))

outliers_log=np.array(outliers_log)
outliers_log=np.unique(outliers_log)
print('Outlier indices log: ',outliers_log)
print('# of outliers log: ',len(outliers_log))
```

```python
spendings_boxcox_IsoFor=pd.read_pickle('DataFrames/spendings_boxcox')
spendings_boxcox_IsoFor.drop(outliers_boxcox,inplace=True)
spendings_boxcox_IsoFor.to_pickle('DataFrames/spendings_boxcox_IsoFor')
```

```python
spendings_log_IsoFor=pd.read_pickle('DataFrames/spendings_log')
spendings_log_IsoFor.drop(outliers_log,inplace=True)
spendings_log_IsoFor.to_pickle('DataFrames/spendings_log_IsoFor')
```
